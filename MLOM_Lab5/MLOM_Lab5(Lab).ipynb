```python
chars=sorted(list(set(raw_text)))
print(chars)
#create mapping

char_to_int = dict((c,i)for i ,c in enumerate(chars)) #crerate a dictonary called char_to_int
```

```python
#Getting the details of the dataset
n_chars=len(raw_text)
n_vocab=len(chars)
print("Total characters:", n_chars)
print("Total vocabs:", n_vocab)
```

```python
#Prepare the dataset of input to output pairs encoded as integers

seq_length = 15
dataX=[]
dataY=[]  #character that have the 16th possition
for i in range(0,n_chars-seq_length, 1):

  seq_in = raw_text[i:i + seq_length]
 # print(seq_in) #this is input
  seq_out = raw_text[i + seq_length]
  #print("this is output")
  dataX.append([char_to_int[char] for char in seq_in])
  dataY.append(char_to_int[seq_out]) #character that can be have in the 16th possition
n_patterns = len(dataY)
print("Total Patterns: " , n_patterns)
 #assign values for each characters as input and they giva a output for that by their own by predicting
  #characters will be assgined by ceratain numbers
```

```python
dataX
```

```python
#Pre Processing task
import numpy as np
X = np.reshape(dataX, (n_patterns, seq_length, 1) )

X = X/ float(n_vocab)
print(X)

#pre processing task


#X = numpy.reshape(dataX, (n_patterns, seq_length, 1))

#rescalling to make it easier to train the model , resclae thevalues  to between 0 and 1
#X= X / float(n_vocab)
#print(X)

#scalling between 0 and 1 beacuse that is the format LSRTM models get the values
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.utils import np_utils

y = np_utils.to_categorical(dataY)

```

```python
from keras.layers.attention.multi_head_attention import activation
from keras.api._v2.keras import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
from keras.layers.attention.multi_head_attention import activation
from keras.api._v2.keras import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
from keras.engine.sequential import Sequential
#define the LSTM model

model = Sequential()
model.add(LSTM(256, input_shape=(X.shape[1], X.shape[2])))
model.add(Dropout(0.2)) #overcome the overfitting
model.add(Dense(y.shape[1], activation = 'softmax')) #softmax wil provide probabilities for each character
model.compile(loss='categorical_crossentropy' , optimizer = 'adam') #categorical crossentropy since we have multiple  classes

#define the checkpoints
filepath ="weights-improvement-{epoch:02d}--{loss:.4f}.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='loss' , verbose=1, save_best_only=True, mode='min')
callbacks_list = [checkpoint]
```

```python
epochs = 10
batch_size = 128
```

```python
model.fit(X, y , epochs = epochs, batch_size=batch_size, callbacks=callbacks_list)
```
